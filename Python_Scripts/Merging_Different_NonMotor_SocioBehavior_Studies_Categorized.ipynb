## Benton_Judgment_of_Line_Orientation

```python
Benton = pd.read_csv("Benton_Judgment_of_Line_Orientation-Archived.csv")
Benton.columns
```

1.1 Removing duplicate patient ID,keeping the first row of each patients

```python
Benton.drop_duplicates(subset = ['PATNO'], keep = 'first', inplace = True)
Benton.head()
```

1.2 Removing not required columns

```python
Benton_cut = Benton.drop(['REC_ID', 'F_STATUS', 'EVENT_ID', 'INFODT', 'BJLOT1', 'BJLOT2', 'BJLOT3', 
                           'BJLOT4', 'BJLOT5', 'BJLOT6', 'BJLOT7',
       'BJLOT8', 'BJLOT9', 'BJLOT10', 'BJLOT11', 'BJLOT12', 'BJLOT13',
       'BJLOT14', 'BJLOT15', 'BJLOT16', 'BJLOT17', 'BJLOT18', 'BJLOT19',
       'BJLOT20', 'BJLOT21', 'BJLOT22', 'BJLOT23', 'BJLOT24', 'BJLOT25',
       'BJLOT26', 'BJLOT27', 'BJLOT28', 'BJLOT29', 'BJLOT30', 'JLO_TOTRAW',
        'AGE_ASSESS_JLO', 'DVS_JLO_MSSA', 'DVS_JLO_MSSAE',
       'ORIG_ENTRY', 'LAST_UPDATE', 'QUERY', 'SITE_APRV'], axis = 1)
Benton_cut.head()
```

1.3 Calculating & Dropping NA values

```python
Benton_cut1 = Benton_cut.dropna()
Benton_cut1.isnull().sum().sum()
Benton_cut1.head()
```

1.4 Renaming a column

```python
Benton_cut1.rename (columns = {'JLO_TOTCALC' : 'Benton_Score30'}, inplace = True)
Benton_cut1.head()
```

25>= 0 = Normal
20>= 1 = Mild
20< 2 = Severe

1.5 Categorizing the values

```python
col         = 'Benton_Score30'
conditions  = [ Benton_cut1[col] >= 25, (Benton_cut1[col] < 25) & (Benton_cut1[col]>= 20), Benton_cut1[col] < 20 ]
choices     = [ 0, 1, 2 ]
    
Benton_cut1["Benton_Category"] = np.select(conditions, choices, default=np.nan)
Benton_cut1.head()
```

1.6 Removing a column

```python
Benton_cut2 = Benton_cut1.drop(["Benton_Score30"], axis = 1)
Benton_cut2.head()
```

```python
Benton_cut2.to_csv("Benton_Judgment_of_Line_Orientation-Categorised.csv")
```
