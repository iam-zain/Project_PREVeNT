# Apathy

```python
%cd Z:\PPMI_Data\Excels\NonMotors\Derived250\GeneCpG\Apathy
df1 = pd.read_csv ("CpG_Apathy_Gene.csv")
df1.head(2)
```

```python
df = pd.merge (df1, Cpg_File, on = 'CpG', how = 'inner')
df.head(2)
```

df.to_csv ("TrailMakingB_CpG_Methylome.csv",index = 'False')

```python
df = df.drop('Gene', axis=1)
```

```python
dft = df.transpose()
dft.head(2)
```

```python
dft.rename(columns=dft.iloc[0], inplace = True)
dft.head(2)
```

```python
dft.index.name ='CpG'
dft.reset_index(inplace=True)
dft.head(2)
```

```python
dft = dft.drop ([0], axis = 0)
dft.head(2)
```

```python
dft.rename (columns = {"CpG" : "Sentrix"}, inplace = True)
dft.head(2)
```

```python
df2 = pd.read_csv ("Apathy_PATNO_APPRDX_Age_in524Methyl.csv")
df3 = pd.merge (df2, dft, on= "Sentrix", how= "inner")
print(df3.shape)
df3.head(2)
```

```python
df3.to_csv ("Apathy_with_MethylomeDetails_Male.csv", index=False)
```
