```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import numpy as np
import joblib
```

```python
%cd Z:\PPMI_Data\Excels\CollaborativeFiltering
NonMot = pd.read_csv('Feats45_unCategAge_APPRDX.csv')
NonMot1 = NonMot.drop(['PATNO', 'Patient_ID', 'Age'], axis = 1)
NonMot1['APPRDX'] = NonMot1['APPRDX'].replace([1], 'Patient')
NonMot1['APPRDX'] = NonMot1['APPRDX'].replace([2], 'Healthy')
NonMot1.head(2)
```

## All features at once

```python
output = []
for i in range(1, 101):
    print(i)
    train_data, test_data, train_labels, test_labels = train_test_split(NonMot1.drop('APPRDX', axis=1), NonMot1['APPRDX'], test_size=0.2)
    
    logreg_model = LogisticRegression()
    parameters = {'C': np.arange(0.1, 10.1, 0.5)}
    logreg_model = GridSearchCV(logreg_model, parameters)
    logreg_model.fit(train_data, train_labels)
    
    predictions = logreg_model.predict(test_data)
    accuracy = accuracy_score(test_labels, predictions)
    print(f'Accuracy: {accuracy}')
    output.append(accuracy)

np.savetxt("NonMotor_unCateg_100LogReg.csv", output, delimiter=",")
```

## Individual Features

```python
for col in NonMot1.columns:
    if col != 'APPRDX':
        output = []
        train_data, test_data, train_labels, test_labels = train_test_split(NonMot1[[col]], NonMot1['APPRDX'], 
                                                                            test_size=0.2, random_state=1)

        logreg_model = LogisticRegression()
        parameters = {'C': np.arange(0.1, 10.1, 0.5)}
        logreg_model = GridSearchCV(logreg_model, parameters)
        logreg_model.fit(train_data, train_labels)

        predictions = logreg_model.predict(test_data)
        accuracy = accuracy_score(test_labels, predictions)
        print(f'Accuracy for column {col}: {accuracy}')
        output.append(accuracy)

        np.savetxt(f"{col}_LogReg.csv", output, delimiter=",")
        joblib.dump(logreg_model, f"{col}_SVM_model.joblib")
```

The range of values for the C hyperparameter in logistic regression problems can vary greatly depending on the specific problem and dataset being used. However, there are some commonly used ranges that work well in many cases.

A common range for C is [0.001, 0.01, 0.1, 1, 10, 100]. This range covers a wide range of values, including values that result in underfitting (small values of C) and overfitting (large values of C).

Another common range is [0.0001, 0.001, 0.01, 0.1, 1, 10]. This range includes smaller values of C than the previous range, which can be useful when the data is very large or there are many features.

It's important to note that the optimal value of C will depend on the specific problem and dataset being used, so it's a good idea to try a range of values and compare the performance of the models. In addition, cross-validation can be used to determine the optimal value of C by evaluating the performance of the model on a validation set.
