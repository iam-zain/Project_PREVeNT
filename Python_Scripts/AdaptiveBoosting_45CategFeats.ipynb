```python
import pandas as pd
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score
import numpy as np
import joblib
```

```python
%cd Z:\PPMI_Data\Excels\CollaborativeFiltering
NonMot = pd.read_csv('Feats45_unCategAge_APPRDX.csv')
NonMot1 = NonMot.drop(['PATNO', 'Patient_ID', 'Age'], axis = 1)
NonMot1['APPRDX'] = NonMot1['APPRDX'].replace([1], 'Patient')
NonMot1['APPRDX'] = NonMot1['APPRDX'].replace([2], 'Healthy')
NonMot1.head(2)
```

```python

output = []
for i in range(1, 101):
    print(i)
    train_data, test_data, train_labels, test_labels = train_test_split(NonMot1.drop('APPRDX', axis=1), 
                                                                        NonMot1['APPRDX'], test_size=0.2)
    
    ada_model = AdaBoostClassifier()
    parameters = {'n_estimators': [200], 'learning_rate': [0.1]}
    ada_model = GridSearchCV(ada_model, parameters)
    ada_model.fit(train_data, train_labels)
    
    predictions = ada_model.predict(test_data)
    accuracy = accuracy_score(test_labels, predictions)
    print(f'Accuracy: {accuracy}')
    output.append(accuracy)

np.savetxt("NonMotor_unCateg_100AdaBoost.csv", output, delimiter=",")
```

The choice of values for the n_estimators and learning_rate hyperparameters depends on the problem you are trying to solve and the size of your data. Here are some common ranges for these hyperparameters:

    n_estimators: The number of weak learners (base estimators) used in the boosting process. A common range for n_estimators is between 50 and 300. Larger values will generally result in a more complex model and a longer training time.

    learning_rate: The learning rate of the algorithm, which controls the contribution of each base estimator to the final result. A common range for learning_rate is between 0.01 and 1. Lower values of learning_rate will result in a slower learning process, but a more accurate model. Higher values of learning_rate will result in a faster learning process, but a less accurate model.

These ranges are just a starting point and may need to be adjusted for your specific problem. The best way to determine the optimal values for these hyperparameters is through experimentation, such as a grid search or a random search.

```python
for col in NonMot1.columns:
    if col != 'APPRDX':
        output = []
        train_data, test_data, train_labels, test_labels = train_test_split(NonMot1[[col]], 
                                                                NonMot1['APPRDX'], test_size=0.2)

        ab_model = AdaBoostClassifier(n_estimators=200, learning_rate = 0.1)
        ab_model.fit(train_data, train_labels)

        predictions = ab_model.predict(test_data)
        accuracy = accuracy_score(test_labels, predictions)
        print(f'Accuracy for column {col}: {accuracy}')
        output.append(accuracy)

        np.savetxt(f"{col}_AdaBoost.csv", output, delimiter=",")
        joblib.dump(ab_model, f"{col}_AdaBoost_model.joblib")
```
