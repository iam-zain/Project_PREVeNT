```python
import pandas as pd
import matplotlib.pyplot as plt

%cd Z:\PPMI_Data\Excels\NonMotors\The251\Female\GeneCpG\Xtras\Top50_eachFeats
```

## Only Patient

```python
data = pd.read_csv('All5_Top50sCpG_Female.csv')
data = data.drop(['PATNO'], axis = 1)
data = data[data['APPRDX'] == 1]
print(data.shape)

# Extract the dependent binary variable and independent columns
dependent_col = data.iloc[:, 0]
independent_cols = data.iloc[:, 1:]

# Calculate the average and standard deviation for each independent column
average_values = independent_cols.mean()
std_deviation_values = independent_cols.std()

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(average_values, std_deviation_values, marker='o', s = 12, color='red')
plt.title('Average vs. Standard Deviation', fontsize=18, fontweight='bold')
plt.xlabel('Average Value', fontsize=14)
plt.ylabel('Standard Deviation', fontsize=14)

# Set axis label fontweight directly
ax = plt.gca()  # Get current axis
ax.xaxis.label.set_fontweight('bold')
ax.yaxis.label.set_fontweight('bold')

# Set tick label fontweight
for tick in ax.xaxis.get_major_ticks():
    tick.label1.set_fontweight('bold')
for tick in ax.yaxis.get_major_ticks():
    tick.label1.set_fontweight('bold')
plt.grid(True)
plt.savefig('average_std_deviation_plot.pdf', format='pdf', bbox_inches='tight', orientation='landscape')
plt.savefig('average_std_deviation_plot.svg', format='svg', bbox_inches='tight', orientation='landscape')
plt.show()

```

## Both Patient and Healthy

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('All5_Top50sCpG_Female.csv')
data = data.drop(['PATNO'], axis=1)

# Separate data based on APPRDX values
data_apprdx_1 = data[data['APPRDX'] == 1]
data_apprdx_2 = data[data['APPRDX'] == 2]

# Calculate average and standard deviation for both cases
avg_std_1 = data_apprdx_1.iloc[:, 1:].agg(['mean', 'std']).transpose()
avg_std_2 = data_apprdx_2.iloc[:, 1:].agg(['mean', 'std']).transpose()

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(avg_std_1['mean'], avg_std_1['std'], marker='o', s = 20, color='red', label='APPRDX=1')
plt.scatter(avg_std_2['mean'], avg_std_2['std'], marker='s', s = 20, color='green', label='APPRDX=2')
plt.title('Average vs. Standard Deviation', fontsize=22, fontweight='bold')
plt.xlabel('Average Value', fontsize=18)
plt.ylabel('Standard Deviation', fontsize=18)
# Make the legend labels bold
legend = plt.legend()
for label in legend.get_texts():
    label.set_fontweight('bold')

# Set axis label fontweight directly
ax = plt.gca()  # Get current axis
ax.xaxis.label.set_fontweight('bold')
ax.yaxis.label.set_fontweight('bold')

# Set tick label fontweight
for tick in ax.xaxis.get_major_ticks():
    tick.label1.set_fontweight('bold')
for tick in ax.yaxis.get_major_ticks():
    tick.label1.set_fontweight('bold')
plt.grid(True)
plt.show()
```
