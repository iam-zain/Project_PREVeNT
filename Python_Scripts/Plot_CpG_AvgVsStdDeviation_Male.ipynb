```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%cd Z:\PPMI_Data\Excels\Raw_And_More
```

```python
Cpg_File = pd.read_csv ("beta_post_Funnorm_PPMI_EPICn524final030618.csv")
Cpg_File.head(3)
```

```python
Cpg_File.index.name ='CpG'
Cpg_File.reset_index(inplace=True)
Cpg_File.head(3)
```

```python
dft = Cpg_File.transpose()
dft.head(2)
```

```python
dft.rename(columns=dft.iloc[0], inplace = True)
dft.head(2)
```

```python
dft.index.name ='CpG'
dft.reset_index(inplace=True)
dft.head(2)
```

```python
dft = dft.drop ([0], axis = 0)
dft.head(2)
```

```python
dft.rename (columns = {"CpG" : "Sentrix"}, inplace = True)
dft.head(2)
```

```python
df1 = pd.read_csv ("Sentrix_APPRDX_of524Methyl_APP1_2_Male.csv")
df1.head(2)
```

```python
df = pd.merge (df1, dft, on = 'Sentrix', how = 'inner')
df.head(2)
```

```python
df = df.drop('Sentrix', axis=1)
```

## Patient

```python
df_P = df[df['APPRDX'] == 1]
print(df.shape)

# Extract the dependent binary variable and independent columns
dependent_col = df_P.iloc[:, 0]
independent_cols = df_P.iloc[:, 1:]

# Calculate the average and standard deviation for each independent column
average_values = independent_cols.mean()
std_deviation_values = independent_cols.std()

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(average_values, std_deviation_values, marker='o', s = 12, color='red')
plt.title('Average vs. Standard Deviation', fontsize=18, fontweight='bold')
plt.xlabel('Average Value', fontsize=14)
plt.ylabel('Standard Deviation', fontsize=14)

# Set axis label fontweight directly
ax = plt.gca()  # Get current axis
ax.xaxis.label.set_fontweight('bold')
ax.yaxis.label.set_fontweight('bold')

# Set tick label fontweight
for tick in ax.xaxis.get_major_ticks():
    tick.label1.set_fontweight('bold')
for tick in ax.yaxis.get_major_ticks():
    tick.label1.set_fontweight('bold')
plt.grid(True)
plt.savefig('CpG_All_average_std_deviation_Male_Diseased.pdf', format='pdf', bbox_inches='tight', orientation='landscape')
plt.savefig('CpG_All_average_std_deviation_Male_Diseased.svg', format='svg', bbox_inches='tight', orientation='landscape')
plt.show()

```

## Both Patient & Healthy

```python

# Separate data based on APPRDX values
data_apprdx_1 = df[df['APPRDX'] == 1]
data_apprdx_2 = df[df['APPRDX'] == 2]

# Calculate average and standard deviation for both cases
avg_std_1 = data_apprdx_1.iloc[:, 1:].agg(['mean', 'std']).transpose()
avg_std_2 = data_apprdx_2.iloc[:, 1:].agg(['mean', 'std']).transpose()

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(avg_std_1['mean'], avg_std_1['std'], marker='o', s = 2, color='red', label='Patient')
plt.scatter(avg_std_2['mean'], avg_std_2['std'], marker='s', s = 2, color='green', label='Healthy')
plt.title('Average vs. Standard Deviation', fontsize=18, fontweight='bold')
plt.xlabel('Average Value', fontsize=14)
plt.ylabel('Standard Deviation', fontsize=14)
# Make the legend labels bold
legend = plt.legend()
for label in legend.get_texts():
    label.set_fontweight('bold')

# Set axis label fontweight directly
ax = plt.gca()  # Get current axis
ax.xaxis.label.set_fontweight('bold')
ax.yaxis.label.set_fontweight('bold')

# Set tick label fontweight
for tick in ax.xaxis.get_major_ticks():
    tick.label1.set_fontweight('bold')
for tick in ax.yaxis.get_major_ticks():
    tick.label1.set_fontweight('bold')
plt.grid(True)
plt.savefig('CpG_All_average_std_deviation_Male.pdf', format='pdf', bbox_inches='tight', orientation='landscape')
plt.savefig('CpG_All_average_std_deviation_Male.svg', format='svg', bbox_inches='tight', orientation='landscape')
plt.show()
```
