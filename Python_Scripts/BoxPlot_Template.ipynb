```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set the working directory
os.chdir("Z:\\PPMI_Data\\Excels\\NonMotors\\The251\\Various_Models")

# Read the CSV file
Non_Mot = pd.read_csv('Various_ML_Model_Compare.csv', encoding='latin1')

# Convert the data from wide to long format
Non_Mot_Long = pd.melt(Non_Mot, id_vars=[], var_name='Feature', value_name='Model_Accuracy')

# Plot a grouped boxplot
sns.set(font_scale=1.2)
g = sns.catplot(x='Feature', y='Model_Accuracy', hue='Feature', data=Non_Mot_Long, kind='box', dodge=False, legend=False)
g.set_xticklabels(rotation=50, ha='right')
g.set_axis_labels('', 'Model Accuracy\n')
g.fig.suptitle('Model Comparison - Patient Vs Healthy Control (Uncategorized)', fontsize=16, fontweight='bold')
plt.show()
```
